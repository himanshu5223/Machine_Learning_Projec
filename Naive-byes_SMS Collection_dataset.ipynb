```python
X = dataset.sms
print(X)
y = dataset.labels
print(y)
```

```python
# splitting into test and train
from sklearn.model_selection  import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print("X_train value:: \n ",X_train)
```

```python
print("y_train value:: \n ",y_train)
```

```python
print("X_test value:: \n ",X_test)
```

```python
print("y_test value:: \n ",y_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
#print(vect)
```

```python
vect.fit(X_train)
```

```python
vect.vocabulary_
```

```python
vect.get_feature_names()
```

```python
X_train_transformed = vect.transform(X_train)
print("X_train_transformed Value :: \n ",X_train_transformed)

X_test_transformed = vect.transform(X_test)
print("X_test transformed Value :: \n ",X_test_transformed)
```

```python
from sklearn.naive_bayes import BernoulliNB
# instantiate bernoulli NB object
bnb = BernoulliNB()

# fit 
bnb.fit(X_train_transformed,y_train)

# predict class
y_pred_class = bnb.predict(X_test_transformed)
print("y_pred_class value ::\n ", y_pred_class)

# predict probability
y_pred_proba = bnb.predict_proba(X_test_transformed)
print("\n y_pred_proba value :: \n", y_pred_proba)

# accuracy
from sklearn import metrics
a = metrics.accuracy_score(y_test, y_pred_class)

print(" \n accuracy of number : \n ", a)
```

```python
from sklearn.metrics import confusion_matrix as sk_confusion_matrix
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred_proba[:,1])
roc_auc = auc(false_positive_rate, true_positive_rate)
```
